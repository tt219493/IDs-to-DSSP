```python
import sys
sys.path.append('/content/ids-to-dssp/src')
```

```python
import ids_to_dssp.processing
import ids_to_dssp.pdb
import ids_to_dssp.rs_dssp
import ids_to_dssp.pipeline
import os
```

```python
input_path = os.path.join('data', 'train.tsv')
dl_path = os.path.join('data', 'pdb')

id_list = ids_to_dssp.processing.ids_to_df(input_path, ids_only = True)
file_list = ids_to_dssp.pdb.ids_to_pdb_download(id_list, dl_path)
```

```python
ss_df = ids_to_dssp.rs_dssp.files_to_dssp(file_list)
```

```python
seq_df = ids_to_dssp.pdb.ids_to_sequences(id_list)
```

```python
full_df = ids_to_dssp.pipeline.combine_to_full(ss_df, seq_df)
```

```python
full_df.collect()
```

```python
ids_to_dssp.processing.df_to_output(full_df, os.path.join('data'), 'train', is_formatted=False, to_parquet=True )
```
